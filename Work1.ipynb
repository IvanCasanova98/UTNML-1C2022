```python
import pandas as pd
import datetime
import numpy as np
FILE_PATH = "data.csv"
```

```python
dataframe = pd.read_csv(FILE_PATH, sep='|', low_memory=False)
```

```python
#Number of Columns
dataframe.shape[1]
```

```python
#Number of Columns from Saving Account
dataframe.filter(regex='SavingAccount_*').shape[1]
```

```python
#Number of Months
dataframe["Month"].nunique()
```

```python
#Number of Clients
dataframe["client_id"].nunique()
```

```python
#Take only clients with 9 Months of Data. Why?
#We take clients with recent activity because those are relevant to our universe definition. 
#If we added clients with less month of data that could affect the model later on.
#We use this amount (9) because of this particular use case but it will always depend in the case at hand.
clients_with_nine_months = dataframe.groupby("client_id").filter(lambda x: len(x.Month.unique()) == 9).client_id
dataframe_with_nine_month_clients = dataframe[dataframe.client_id.isin(clients_with_nine_months)]

dataframe_with_nine_month_clients
```

```python
#Take only clients without Active Package. Be careful with the month! 
dates = pd.to_datetime(dataframe_with_nine_month_clients['Month'], format='%Y-%m-%d')
last_date = dates.max().strftime('%Y-%m-%d')
last_date_clients_with_no_package = dataframe_with_nine_month_clients[(dataframe_with_nine_month_clients.Month == last_date) & (dataframe_with_nine_month_clients.Package_Active == 'No')].client_id
dataframe_with_nine_month_clients_and_no_package = dataframe_with_nine_month_clients[dataframe_with_nine_month_clients.client_id.isin(last_date_clients_with_no_package)]

dataframe_with_nine_month_clients_and_no_package
```

```python
TOTAL_MONTH_RANGE = 9

PREDICTION_WINDOW_RANGE = 2
LEAD_WINDOW_RANGE = 1
HISTORICAL_WINDOW_RANGE = TOTAL_RANGE - PREDICTION_WINDOW_RANGE - LEAD_WINDOW_RANGE

#First i convert dates to type date so i can order them
dates = pd.to_datetime(dataframe['Month'], format='%Y-%m-%d').unique()[:TOTAL_MONTH_RANGE]
dates =  np.datetime_as_string(np.sort(dates), unit='D')
dates
```

```python
prediction_window_data = dataframe_with_nine_month_clients_and_no_package[dataframe_with_nine_month_clients_and_no_package.Month.isin(dates[-PREDICTION_WINDOW_RANGE:])]
prediction_window_data
```

```python
lead_window_data = dataframe_with_nine_month_clients_and_no_package[dataframe_with_nine_month_clients_and_no_package.Month.isin([dates[-PREDICTION_WINDOW_RANGE-LEAD_WINDOW_RANGE]])]
lead_window_data
```

```python
historical_window_data =  dataframe_with_nine_month_clients_and_no_package[dataframe_with_nine_month_clients_and_no_package.Month.isin(dates[:HISTORICAL_WINDOW_RANGE])]
historical_window_data
```

```python
#Is the target Balanced or Unbalanced?
targets_data = dataframe_with_nine_month_clients_and_no_package[dataframe_with_nine_month_clients_and_no_package.Target == 1.0]
targets_count = targets_data.Target.count()

total_count = dataframe_with_nine_month_clients_and_no_package.Target.count()

balance_percentage = targets_count/total_count

balance_percentage 
```
